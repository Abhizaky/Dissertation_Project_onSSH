```python
from sklearn.decomposition import PCA
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import ConstantKernel, RBF, WhiteKernel
```

```python
arr = ds.ssh.isel(t=slice(0,50)).load().values   
nt, ny, nx = arr.shape
print(arr)
```

```python
import numpy as np
lookback = 5
X, y = [], []
for i in range(nt - lookback):
    X.append(arr[i:i+lookback])
    y.append(arr[i+lookback])
X = np.array(X)               # (45, lookback, ny, nx)
y = np.array(y)  
```

```python
# Flatten spatial dims
X_flat = X.reshape(X.shape[0], -1)   # (45, lookback*ny*nx)
y_flat = y.reshape(y.shape[0], -1)   # (45, ny*nx)
```

```python
n_components = 10
pca = PCA(n_components=n_components)
y_pca = pca.fit_transform(y_flat)
```

```python
kernel = ConstantKernel(1.0) * RBF(length_scale=1.0) + WhiteKernel(noise_level=1.0)
gp = GaussianProcessRegressor(kernel=kernel, normalize_y=True)
```

```python
gp.fit(X_flat, y_pca)
```

```python
# 5. Predict the next SSH map
last_window = arr[-lookback:]                  # (lookback, ny, nx)
last_flat   = last_window.reshape(1, -1)
```

```python
y_pca_pred = gp.predict(last_flat)             # (1, 10)
y_flat_pred = pca.inverse_transform(y_pca_pred)  # (1, ny*nx)
next_ssh    = y_flat_pred.reshape(ny, nx) 
```

```python
print("Predicted next SSH frame shape:", next_ssh.shape)
```
